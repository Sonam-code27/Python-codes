## Dimensionality Reduction with PCA

```python
# Standardize the features to ensure they have a mean of 0 and standard deviation of 1
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Standardizing the dataset
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df)
```

```python
# Chwck the mean and standard devaition of scaled data
scaled_data.mean(axis=0),scaled_data.std(axis=0)
```

```python
# Implement PCA to reduce the diamensionality of the data set
```

```python
from sklearn.decomposition import PCA
```

```python
# Apply PCA
pca=PCA()
pca.fit(scaled_data)
```

```python
#Explained variance ratio for each principal component
explained_variance_ratio=pca.explained_variance_ratio_
```

```python
#Plot the explained variance
plt.figure(figsize=(8,6))
plt.plot(range(1,len(explained_variance_ratio)+1),explained_variance_ratio,marker='o')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

```python
cumulative_variance=pca.explained_variance_ratio_.cumsum()
```

```python
# plot the cumulative explained variance
```

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
# Transform the original dataset into the principal components
```

```python
n_components=next(i for i,v in enumerate(cumulative_variance) if v>=0.95)+1
```

```python
pca=PCA(n_components=n_components)
pca_data=pca.fit_transform(scaled_data)
```

```python
pca_data
```

## Clustering with original data

```python
from sklearn.cluster import KMeans
n_clusters=3
kmeans=KMeans(n_clusters=n_clusters)
kmeans.fit(scaled_data)
```

```python
labels=kmeans.labels_
```

```python
# Visualize the clustering result
# Scatter plot for clustering visualization
plt.figure(figsize=(8,6))
sns.scatterplot(x=scaled_data[:,0],y=scaled_data[:,1],hue=labels,palette='viridis',s=100,alpha=0.7)
plt.title('K-means Clustering on Original Data')
plt.show()
```

```python
# Evaluate the clustering performnace using metrics
```

```python
from sklearn.metrics import silhouette_score,davies_bouldin_score
sil_score=silhouette_score(scaled_data,labels)
print(f'Silhouette Score:{sil_score}')
```

```python
db_index=davies_bouldin_score(scaled_data,labels)
print(f'Davies_Bouldin Index:{db_index}')
```

## Clustering with PCA data

```python
# Apply K_means on PCA transfromed data
```

```python
kmeans_pca=KMeans(n_clusters=n_clusters)
kmeans_pca.fit(pca_data)
```

```python
labels_pca=kmeans_pca.labels_
```

```python
# Visualize the clustering result from PCA
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_data[:,0],y=pca_data[:,1],hue=labels_pca,palette='viridis',s=100,alpha=0.7)
plt.title('K-Means Clustering on PCA Data')
plt.show()
```

```python
sil_score=silhouette_score(pca_data,labels)
print(f'Silhouette Score:{sil_score}')
```

```python
db_index=davies_bouldin_score(pca_data,labels)
print(f'Davies_Bouldin Index:{db_index}')
```

```python
# Compare clustering result
```

```python
#Visual Comparision of the clustering results using subplots
```

```python
fig,axes=plt.subplots(1,2,figsize=(16,6))
# Original Data Clustering
axes[0].scatter(scaled_data[:,0],scaled_data[:,1],c=labels,cmap='viridis',s=100,alpha=0.7)
axes[0].set_title('Original Data Clustering')
#PCA data Clustering
axes[1].scatter(pca_data[:,0],pca_data[:,1],c=labels_pca,cmap='viridis',s=100,alpha=0.7)
axes[1].set_title('PCA Data Clustering')
plt.show()
```
